## **Stemming**
Stemming reduces words to their root form.

Stemming is an important text-processing technique that reduces words to their base or root form by removing prefixes and suffixes. This process standardizes words which helps to improve the efficiency and effectiveness of various natural language processing (NLP) tasks.

**Examples of stemming for the word "like":**

"likes" → "like"

"liked" → "like"

"likely" → "like"

"liking" → "like"

```python
from nltk.stem import PorterStemmer
porter_stemmer = PorterStemmer()
words = ["running", "jumps", "liked", "running", "playing","happily"]
stemmed_words = [porter_stemmer.stem(word) for word in words]
print("Original words:", words)
print("Stemmed words:", stemmed_words)

#Snowball Stemmer : is an enhanced version of the Porter Stemmer,Supports multiple languages.
from nltk.stem import SnowballStemmer
stemmer = SnowballStemmer(language='english')
words_to_stem = ['running', 'jumped', 'happily', 'quickly', 'foxes']
stemmed_words = [stemmer.stem(word) for word in words_to_stem]
print("Original words:", words_to_stem)
print("Stemmed words:", stemmed_words)

#Lancaster Stemmer: is known for being more aggressive and faster than other stemmers.
from nltk.stem import LancasterStemmer
stemmer = LancasterStemmer()
words_to_stem = ['running', 'jumped', 'happily', 'quickly', 'foxes']
stemmed_words = [stemmer.stem(word) for word in words_to_stem]
print("Original words:", words_to_stem)
print("Stemmed words:", stemmed_words)
```

## **Lemmatization**

Lemmatization produces meaningful root words.

**Example**

better → good

running → run

Lemmatization is an important text pre-processing technique in Natural Language Processing (NLP) that reduces words to their base form known as a "lemma." For example, the lemma of "running" is "run" and "better" becomes "good."

Unlike stemming which simply removes prefixes or suffixes, it considers the word's meaning and part of speech (POS) and ensures that the base form is a valid word. This makes lemmatization more accurate as it avoids generating non-dictionary words.

**Used for:**

Improves accuracy

Reduced Data Redundancy

Better NLP Model Performance




```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

print(lemmatizer.lemmatize("running", pos="v"))

print(lemmatizer.lemmatize("better", pos="a"))
```

## **Bag of Words (BoW)**

BoW represents text as word frequency vectors.

In Natural Language Processing (NLP) text data needs to be converted into numbers so that machine learning algorithms can understand it. One common method to do this is Bag of Words (BoW) model. It turns text like sentence, paragraph or document into a collection of words and counts how often each word appears but ignoring the order of the words. It does not consider the order of the words or their grammar but focuses on counting how often each word appears in the text.

This makes it useful for tasks like text classification, sentiment analysis and clustering.

### **Step 1: Preprocessing the Text**

Clean the text

✔ Convert to lowercase

✔ Remove special characters

```python
import re

dataset = []
text = [
    "I love coding!",
    "I love Python$",
    "Python coding is   fun"
]

for sentence in text:
    sentence = sentence.lower()              # lowercase
    sentence = re.sub(r'\W', ' ', sentence)  # remove symbols
    sentence = re.sub(r'\s+', ' ', sentence) # remove extra spaces
    dataset.append(sentence)

print(dataset)
```

### **Step 2: Counting Word Frequencies**

Count how many times each word appears

```python
from collections import defaultdict

word2count = defaultdict(int)

for sentence in dataset:
    words = sentence.split()
    for word in words:
        word2count[word] += 1

print(word2count)
```

### **Step 3: Selecting Most Frequent Words**

Let’s take top 3 words

```python
import heapq

freq_words = heapq.nlargest(3, word2count, key=word2count.get)

print(freq_words)
```

### **Step 4: Build Bag of Words Matrix**

Now we create a matrix:

👉 Rows = sentences

👉 Columns = frequent words

👉 Value = 1 (present) or 0 (absent)

```python
X = []

for sentence in dataset:
    vector = []
    words = sentence.split()

    for word in freq_words:
        if word in words:
            vector.append(1)
        else:
            vector.append(0)

    X.append(vector)

print(X)
```

# **N-grams**

N-grams capture word sequences.

N-gram is a technique that splits text into sequences of N consecutive words to capture context and word order for machine learning models.

N-gram = group of N consecutive words

1 word → Unigram

2 words → Bigram

3 words → Trigram

**🧠 Key Idea**

👉 N-grams capture word relationships

BoW → "love", "coding"

Bigram → "love coding" ✔ (more meaningful)

#### **✅ Simple Example**

Sentence:

#### "I love coding"
####🔹 Unigrams (N = 1)
["I", "love", "coding"]
####🔹 Bigrams (N = 2)
["I love", "love coding"]
####🔹 Trigrams (N = 3)
["I love coding"]

```python
#Step 1: Preprocessing Text
import re

text = [
    "I love coding",
    "I love Python",
    "Python coding is fun"
]

dataset = []

for sentence in text:
    sentence = sentence.lower()
    sentence = re.sub(r'\W', ' ', sentence)
    sentence = re.sub(r'\s+', ' ', sentence)
    dataset.append(sentence)

print(dataset)
```

```python
#🟢 Step 2: Create N-grams
#👉 Function to generate N-grams
def generate_ngrams(sentence, n):
    words = sentence.split()
    ngrams = []

    for i in range(len(words) - n + 1):
        ngrams.append(" ".join(words[i:i+n]))

    return ngrams
```

```python
#Example: Bigrams (N = 2)
for sentence in dataset:
    print(generate_ngrams(sentence, 2))

#Example: Trigrams (N = 3)
for sentence in dataset:
    print(generate_ngrams(sentence, 3))
```

```python
#Step 3: Count N-gram Frequencies
from collections import defaultdict

ngram_count = defaultdict(int)

for sentence in dataset:
    ngrams = generate_ngrams(sentence, 2)  # bigrams
    for gram in ngrams:
        ngram_count[gram] += 1

print(ngram_count)
```

```python
#Step 4: Convert into Feature Vector (like BoW)
#Select top N-grams:
import heapq

top_ngrams = heapq.nlargest(3, ngram_count, key=ngram_count.get)
print(top_ngrams)
```

```python
#👉 Build Matrix
X = []

for sentence in dataset:
    vector = []
    sentence_ngrams = generate_ngrams(sentence, 2)

    for gram in top_ngrams:
        if gram in sentence_ngrams:
            vector.append(1)
        else:
            vector.append(0)

    X.append(vector)

print(X)
```

**🤔 Why Do We Need TF-IDF?**

Problem with BoW:

Words like "the", "is", "I" appear many times

But they are not important

👉 BoW treats all words equally ❌

**✅ TF-IDF solves this by:**

Giving high weight to important words

Giving low weight to common words

Sentences:

"I love python"

"Python is powerful"

# **🎯 Word Embeddings:**

👉 Word Embeddings = representing words as meaningful numbers (vectors)

Unlike BoW/TF-IDF:

They don’t just count words ❌

They capture meaning and relationships ✔️

🤔 Why Do We Need Word Embeddings?
❌ Problem with TF-IDF / BoW

**Example:**

"good" and "excellent"

👉 BoW / TF-IDF:

Treat them as completely different ❌

**✅ Word Embeddings Fix This**

👉 They understand:

"good" ≈ "excellent" ✔

"king" - "man" + "woman" ≈ "queen" 👑

### **🧠 Key Idea**

👉 Each word is converted into a vector (list of numbers)

**Example:**

"king"   → [0.2, 0.8, 0.1]

"queen"  → [0.25, 0.75, 0.15]

👉 Similar words → similar vectors ✔

### **Types of Word Embeddings**
**1. Word2Vec :**

👉 Word2Vec is a technique to create word embeddings (vectors) by learning from context.

**Idea:**

“Words that appear in similar contexts have similar meanings”

**Two methods:**

CBOW (predict word from context)

Skip-gram (predict context from word)

### **🔵 1. CBOW (Continuous Bag of Words)**
🎯 Idea:

👉 Predict the center word using surrounding words

📌 Example

Sentence:

"I love coding"

👉 Input (context words):

"I" and "coding"

👉 Output (target word):

"love"

### **🔴 2. Skip-gram**

🎯 Idea:

👉 Predict surrounding words using center word

📌 Example

Sentence:

"I love coding"

👉 Input (center word):

"love"

👉 Output (context words):

"I", "coding"

```python
!pip install gensim
```

```python
# CBOW
from gensim.models import Word2Vec

# Training data (sentences)
sentences = [
    ["i", "love", "coding"],
    ["i", "love", "python"],
    ["python", "is", "powerful"],
    ["coding", "is", "fun"]
]

# Train CBOW model (sg=0)
cbow_model = Word2Vec(
    sentences,
    vector_size=5,   # size of word vectors
    window=2,        # context window
    min_count=1,
    sg=0             # 0 = CBOW
)

# Get word vector
print("Vector for 'coding':")
print(cbow_model.wv["coding"])

# Find similar words
print("\nWords similar to 'coding':")
print(cbow_model.wv.most_similar("coding"))
```

```python
#Skip-Gram
from gensim.models import Word2Vec

# Training data
sentences = [
    ["i", "love", "coding"],
    ["i", "love", "python"],
    ["python", "is", "powerful"],
    ["coding", "is", "fun"]
]

# Train Skip-gram model (sg=1)
skipgram_model = Word2Vec(
    sentences,
    vector_size=5,
    window=2,
    min_count=1,
    sg=1             # 1 = Skip-gram
)

# Get word vector
print("Vector for 'coding':")
print(skipgram_model.wv["coding"])

# Find similar words
print("\nWords similar to 'coding':")
print(skipgram_model.wv.most_similar("coding"))
```

## **🎯 What is NER?**

👉 Named Entity Recognition (NER) is the process of identifying and classifying important words in a sentence into categories like:

Person 👤

Location 🌍

Organization 🏢

Date 📅

Money 💰

**Example**

Sentence:

"Elon Musk founded Tesla in California in 2003"

NER Output:

Elon Musk  → PERSON  
Tesla      → ORGANIZATION  
California → LOCATION  
2003       → DATE  

```python
sentence = "Barack Obama visited India"

words = nltk.word_tokenize(sentence)
pos_tags = nltk.pos_tag(words)

ner_tree = nltk.ne_chunk(pos_tags)

print(ner_tree)

#GPE = Geo-Political Entity (country/place)
```
