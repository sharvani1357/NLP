Task 1 — Text Cleaning System

```python
import re
import string
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
text = "Hello!!! I am learning NLP @ CVR College :)"
text = text.lower()
text = text.translate(str.maketrans('','',string.punctuation))
words = text.split()
stop_words = set(stopwords.words('english'))
cleaned_words = []
for word in words:
    if word not in stop_words:
        cleaned_words.append(word)
cleaned_text = " ".join(cleaned_words)
print(cleaned_text)
```

Task 2 — Tokenization Analyzer

```python
import nltk
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
nltk.download('all')
text = """Natural Language Processing is interesting.It helps computers understand text.
NLP is widely used in AI applications."""
sentences = sent_tokenize(text)
words = word_tokenize(text)
unique_words = set(words)
print("Total Sentences:")
print(len(sentences))
print()
print("Total Words:")
print(len(words))
print()
print("Unique Words:")
print(len(unique_words))
print()
print("Sentences:")
print(sentences)
print()
print("Words:")
print(words)
```

Task 3 — Stopword Removal Tool

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
text = "This is an amazing NLP session for the students"
words = text.split()
stop_words = set(stopwords.words('english'))
filtered_words = []
for word in words:
    if word.lower() not in stop_words:
        filtered_words.append(word)
output = " ".join(filtered_words)
print(output)
```

Task 4 — Stemming vs Lemmatization Comparison

```python
import nltk
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
text = "running studies caring"
words = text.split()
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
stemmed_words = []
lemmatized_words = []
for word in words:
  stemmed_words.append(stemmer.stem(word))
  lemmatized_words.append(lemmatizer.lemmatize(word))
print("Stemmed Words:")
print(stemmed_words)
print()
print("Lemmatized Words:")
print(lemmatized_words)
```

Task 5 — Bag of Words Implementation

```python
from sklearn.feature_extraction.text import CountVectorizer
documents = ["I love NLP","NLP is interesting","I love Machine Learning"]
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(documents)
print("Vocabulary:")
print(vectorizer.vocabulary_)
print()
print("Document Vectors:")
print(X.toarray())
```

Task 6 — TF-IDF Based Important Word Detection

```python
from sklearn.feature_extraction.text import TfidfVectorizer
documents = ["NLP is used in search engines","Search engines use TF IDF","TF IDF helps document ranking"]
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(documents)
feature_names = vectorizer.get_feature_names_out()
print("Important Words:")
print(feature_names)
print()
print("TF IDF Matrix:")
print(X.toarray())
```
